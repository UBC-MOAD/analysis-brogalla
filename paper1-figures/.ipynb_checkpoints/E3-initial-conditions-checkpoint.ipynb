# Initial conditions

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.basemap import Basemap, cm
import cmocean
import netCDF4 as nc
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

##### Set parameters:

```python
land_color = '#a9a7a2'
```

##### Load files:

```python
# Initial conditions from Marco van Hulten
folder = '/ocean/brogalla/GEOTRACES/data/van_hulten/'
file   = 'Fmang-16B05-LD40_Bio_1y_ptrc_T_P80.nc' 

a      = nc.Dataset(folder+file)
mn_dis = np.array(a.variables['dMn']) # time, deptht, y, x
depths = np.array(a.variables['deptht'])
lats   = np.array(a.variables['nav_lat'])
lons   = np.array(a.variables['nav_lon'])
```

```python
# Meshmasks
mask      = nc.Dataset('/ocean/brogalla/GEOTRACES/ariane_runs/ANHA12_Ariane_mesh.nc')
tmask     = np.array(mask.variables['tmask'])
land_mask = np.ma.masked_where((tmask[0,:,:,:] > 0.1) , tmask[0,:,:,:]) 

mesh  = nc.Dataset('/data/brogalla/old/meshmasks/ANHA12_mesh1.nc')
lon   = np.array(mesh.variables['nav_lon'])
lat   = np.array(mesh.variables['nav_lat'])
```

```python
ini_file = nc.Dataset('/ocean/brogalla/GEOTRACES/data/Mn_ini.nc')
ini_lons = np.array(ini_file.variables['nav_lon'])
ini_lats = np.array(ini_file.variables['nav_lat'])
ini_dmn  = np.array(ini_file.variables['dissolmn'])
```

##### Figures:

```python
fig, ax = plt.subplots(1,1, figsize=(10,6))
proj1 = Basemap(projection='stere',lat_0 = 50, lon_0 = -180, resolution='i', width=9.0e6, height=7.0e6, ax=ax)
proj1.drawcoastlines(color='#373736', linewidth=1.8,zorder=2)
proj1.fillcontinents(color=land_color, lake_color=land_color, zorder=2);

x, y = proj1(lons, lats)

CB = proj1.contourf(x, y, mn_dis[0,0,:,:])#, vmin=0, vmax=14e-9, extend='both')
#                          levels=np.arange(vmin_glac, vmax_glac+step_glac, step_glac), \
#                          vmin=vmin_glac, vmax=vmax_glac, cmap=glac_cmap, extend='both')

cbaxes1 = fig.add_axes([0.10, 0.15, 0.03, 0.5]) 
CBar1   = plt.colorbar(CB, ax=ax, cax=cbaxes1, ticks=[2e-9, 4e-9, 6e-9, 8e-9, 10e-9, 12e-9, 14e-9], drawedges=True)  
CBar1.ax.tick_params(axis='y', length=0, labelsize=12, pad=2)
```
