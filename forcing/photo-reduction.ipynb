# Create forcing files

Read in monthly icemod files:

```python
def load_variables(filename):
    a = nc.Dataset(filename)
    
    sol_flux = np.array(a.variables['iocesflx'][:,0,:,:])     
    
    return sol_flux
```

```python
def red_rate_calc(year):
    sol_flux = load_variables('/data/brogalla/ANHA12/icemod_ave/ANHA12_EXH006_'+str(year)+'_monthly.nc')
    red_rate = (np.abs((sol_flux - sol_flux[:,1480:2180,160:800].mean()))/sol_flux[:,1480:2180,160:800].max())*58
    red_rate[red_rate < 1] = 1
    
    return red_rate
```

```python
file_year=2018
```

```python
red_rateyear = red_rate_calc(file_year)
```

```python
print(np.amax(red_rateyear[:,1480:2180,160:800]), np.amin(red_rateyear[:,1480:2180,160:800]))
```

Save files

```python
location='/data/brogalla/reduction_forcing/'
```

```python
def save_file(filename, field1):
    ncd = nc.Dataset(filename, 'w', zlib=True)

    ncd.createDimension('x',len(mesh.dimensions['x']))
    ncd.createDimension('y',len(mesh.dimensions['y']))
    ncd.createDimension('time_counter',None)
    
    # variables
    red             = ncd.createVariable('kred', 'float64', ('y','x'))
    red.units       = 'Unitless'
    red.long_name   = 'Photoreduction multiplier'  
    red.coordinates = 'nav_lon nav_lat'
    red[:]          = field1[:]

    ncd.close()
    return
```

```python
for i in range(1,13):
    if i < 10:
        save_file(location+'red_y'+str(file_year)+'m0'+str(i)+'.nc',red_rateyear[i-1,:,:])
    else:
        save_file(location+'red_y'+str(file_year)+'m'+str(i)+'.nc',red_rateyear[i-1,:,:])
```
