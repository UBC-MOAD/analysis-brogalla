##### Calculations:

```python
mn_m0     = nc.Dataset(initial)
mn_m1     = nc.Dataset(end_yr1 + monthly)
mn_m2     = nc.Dataset(end_yr2 + monthly)
mn_m3     = nc.Dataset(end_yr3 + monthly)
mn_m4     = nc.Dataset(end_yr4 + monthly)
mn_m5     = nc.Dataset(end_yr5 + monthly)
mn_m6     = nc.Dataset(end_yr6 + monthly)
mn_m7     = nc.Dataset(end_yr7 + monthly)
mn_m8     = nc.Dataset(end_yr8 + monthly)
```

```python
dmn_month1 = np.array(mn_m1.variables['dissolmn'][:,0,:,:,:])
dmn_month2 = np.array(mn_m2.variables['dissolmn'][:,0,:,:,:])
dmn_month3 = np.array(mn_m3.variables['dissolmn'][:,0,:,:,:])
dmn_month4 = np.array(mn_m4.variables['dissolmn'][:,0,:,:,:])
dmn_month5 = np.array(mn_m5.variables['dissolmn'][:,0,:,:,:])
dmn_month6 = np.array(mn_m6.variables['dissolmn'][:,0,:,:,:])
dmn_month7 = np.array(mn_m7.variables['dissolmn'][:,0,:,:,:])
dmn_month8 = np.array(mn_m8.variables['dissolmn'][:,0,:,:,:])
```

```python
omn_month1 = np.array(mn_m1.variables['oxidismn'][:,0,:,:,:])
omn_month2 = np.array(mn_m2.variables['oxidismn'][:,0,:,:,:])
omn_month3 = np.array(mn_m3.variables['oxidismn'][:,0,:,:,:])
omn_month4 = np.array(mn_m4.variables['oxidismn'][:,0,:,:,:])
omn_month5 = np.array(mn_m5.variables['oxidismn'][:,0,:,:,:])
omn_month6 = np.array(mn_m6.variables['oxidismn'][:,0,:,:,:])
omn_month7 = np.array(mn_m7.variables['oxidismn'][:,0,:,:,:])
omn_month8 = np.array(mn_m8.variables['oxidismn'][:,0,:,:,:])
```

The year-to-year change is calculated as the sum over all depths of the concentration at the observed stations, $stn$, for each month from one year compared to the previous year. 

$$ diff_{stn} = \sum^{50}_{z=0} | Mn_{stn}(z,month,yr) - Mn_{stn}(z,month,yr-1) | $$

```python
ddiff1 = np.array([np.sum(np.abs(dmn_month2[:,:,i,j]-dmn_month1[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff2 = np.array([np.sum(np.abs(dmn_month3[:,:,i,j]-dmn_month2[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff3 = np.array([np.sum(np.abs(dmn_month4[:,:,i,j]-dmn_month3[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff4 = np.array([np.sum(np.abs(dmn_month5[:,:,i,j]-dmn_month4[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff5 = np.array([np.sum(np.abs(dmn_month6[:,:,i,j]-dmn_month5[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff6 = np.array([np.sum(np.abs(dmn_month7[:,:,i,j]-dmn_month6[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
ddiff7 = np.array([np.sum(np.abs(dmn_month8[:,:,i,j]-dmn_month7[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
```

```python
odiff1 = np.array([np.sum(np.abs(omn_month2[:,:,i,j]-omn_month1[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff2 = np.array([np.sum(np.abs(omn_month3[:,:,i,j]-omn_month2[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff3 = np.array([np.sum(np.abs(omn_month4[:,:,i,j]-omn_month3[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff4 = np.array([np.sum(np.abs(omn_month5[:,:,i,j]-omn_month4[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff5 = np.array([np.sum(np.abs(omn_month6[:,:,i,j]-omn_month5[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff6 = np.array([np.sum(np.abs(omn_month7[:,:,i,j]-omn_month6[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
odiff7 = np.array([np.sum(np.abs(omn_month8[:,:,i,j]-omn_month7[:,:,i,j]), axis=1) for i, j in zip(stn_i, stn_j)])
```

##### Figures:

Reference run:

```python
fig, ax = plt.subplots(1, 1, figsize=(12,5))

nruns  = 7 # number of runs
months = range(12 +1, 12*nruns +1)
years  = np.linspace(2, 2+nruns, nruns*12 + 1)[0:-1]
differences = np.empty((len(name),len(years)))

for stn in range(0,len(name)):
    differences[stn,:] = np.concatenate([ddiff1[stn,:], ddiff2[stn,:], ddiff3[stn,:], ddiff4[stn,:], \
                                         ddiff5[stn,:], ddiff6[stn,:], ddiff7[stn,:]])
    ax.plot(years, differences[stn,:], c=colors[stn], label=name[stn], linewidth=2)
    ax.plot(years, differences[stn,:], '.', c=colors[stn], markersize=10)
    
# Line for spread in values
ax.plot(years, np.max(differences,axis=0)-np.min(differences,axis=0), '-k', linewidth=5, label='spread')

ax.set_xlim([1.9, 2+nruns])
ax.set_ylim([0, 3.e-8])
ax.xaxis.set_ticks(np.linspace(2,2+nruns, nruns*2 + 1))
ax.tick_params(axis='both', which='major', labelsize=10)
ax.set_xlabel('Years of spin up', fontsize=12)
ax.set_ylabel(r'dMn change w.r.t. previous year [mol L$^{-1}$]', fontsize=12)
font = matplotlib.font_manager.FontProperties(family='Helvetica', size=10)
ax.legend(bbox_to_anchor=(0.88, 0.2), frameon=False, prop=font)
```
