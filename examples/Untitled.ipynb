```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
EOSC111_2018 = [4.4, 4.2, 4.0, 4.6, 4.3, 4.4]
EOSC111_2017 = [3.7, 3.6, 3.6, 4.0, 4.1, 3.8]
labelsx = ['','UMI1','UMI2','UMI3','UMI4','UMI5','UMI6']
labelsy = ['','Strongly \n disagree','Disagree','Neutral','Agree','Strongly \n agree']
```
