# Map of sub-domain of ANHA12

```python
import matplotlib.pyplot as plt
import netCDF4 as nc
import numpy as np
from mpl_toolkits.basemap import Basemap, cm
import cmocean
import pickle

%matplotlib inline
```

###### Parameters:

```python
# domain dimensions:
imin, imax = 1100, 2180
jmin, jmax = 160, 800

# Resolution of grid cells to show in figure:
grid_reso = 10
```

###### Load files:

```python
mask     = nc.Dataset('/ocean/brogalla/GEOTRACES/data/ANHA12/ANHA12_mask_Pb.nc')
mesh_lon = np.array(mask.variables['nav_lon'])
mesh_lat = np.array(mask.variables['nav_lat'])
mask = np.array(mask.variables['tmask'][0])
```

```python
data = nc.Dataset('/data/brogalla/run_storage/spin-up0-pb-test/ANHA12_EXH006_5d_20020101_20021231_ptrc_T_20020101-20020105.nc')
dpb = np.array(data.variables['dissolpb'])[0,:,:,:]
dpb_masked = np.ma.masked_where((mask[:,imin:imax,jmin:jmax] == 0), dpb)
```

###### Figure:

```python
fig, ax1, proj1 = pickle.load(open('/ocean/brogalla/GEOTRACES/pickles/pb-surface.pickle','rb'))

x, y = proj1(mesh_lon[imin:imax,jmin:jmax], mesh_lat[imin:imax,jmin:jmax]) 
proj1.contourf(x, y, dpb[0,:,:], levels=20, vmin=0, vmax = 30e-12, extend='both')

# Color bar: --------------------------------------------------------------------------------
# cbaxes = fig.add_axes([0.92, 0.155, 0.02, 0.38]);
# CBar   = plt.colorbar(bath, ax=ax1, cax=cbaxes); 
# CBar.set_label('Ocean depth [m]', fontsize=14)
# CBar.ax.tick_params(axis='y', length=0, labelsize=14)
```
