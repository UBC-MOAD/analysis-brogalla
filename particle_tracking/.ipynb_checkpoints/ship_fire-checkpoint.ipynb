# Track what happens if ship on fire releases materials

```python
import numpy as np
from datetime import timedelta as delta
import netCDF4 as nc
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt
import xarray as xr
from parcels import FieldSet, ParticleSet, JITParticle, AdvectionRK4_3D, ErrorCode, AdvectionDiffusionEM, AdvectionDiffusionM1

%matplotlib inline
```

Load SalishSeaCast ocean model fields:

```python
nowcast  = '/results/SalishSea/nowcast-blue.201905/24oct21/'
forecast = '/results/SalishSea/forecast.201905/24oct21/'
land_color = '#a9a7a2'
```

```python
mesh_mask = '/ocean/brogalla/GEOTRACES/data/mesh_mask201702.nc'

nowcastU  = f'{nowcast}SalishSea_1h_20211024_20211024_grid_U.nc'
nowcastV  = f'{nowcast}SalishSea_1h_20211024_20211024_grid_V.nc'
nowcastW  = f'{nowcast}SalishSea_1h_20211024_20211024_grid_W.nc'

forecastU = f'{forecast}SalishSea_1h_20211025_20211026_grid_U.nc'
forecastV = f'{forecast}SalishSea_1h_20211025_20211026_grid_V.nc'
forecastW = f'{forecast}SalishSea_1h_20211025_20211026_grid_W.nc'

files_U = [nowcastU, forecastU]
files_V = [nowcastV, forecastV]
files_W = [nowcastW, forecastW]
```

## Run forward particle tracking

```python
filenames = {'U': {'lon': mesh_mask, 'lat': mesh_mask, 'depth': files_W[0], 'data': files_U},
             'V': {'lon': mesh_mask, 'lat': mesh_mask, 'depth': files_W[0], 'data': files_V},
             'W': {'lon': mesh_mask, 'lat': mesh_mask, 'depth': files_W[0], 'data': files_W}}

dimensions = {'U': {'lon': 'glamf', 'lat': 'gphif', 'depth': 'depthw', 'time': 'time_counter'},
              'V': {'lon': 'glamf', 'lat': 'gphif', 'depth': 'depthw', 'time': 'time_counter'},
              'W': {'lon': 'glamf', 'lat': 'gphif', 'depth': 'depthw', 'time': 'time_counter'}}

variables = {'U': 'vozocrtx',
             'V': 'vomecrty',
             'W': 'vovecrtz'}

fieldset  = FieldSet.from_nemo(filenames, variables, dimensions)
```

```python
def DeleteParticle(particle, fieldset, time):
    print("Particle [%d] lost !! (%g %g %g %g)" % (particle.id, particle.lon, particle.lat, particle.depth, particle.time))
    particle.delete()
```

```python
release_lon   = -123.36337 # east
release_lat   = 48.33805   # north
release_depth = 1          #

pset = ParticleSet.from_list(fieldset = fieldset,       # the fields on which the particles are advected
                             pclass   = JITParticle,    # the type of particles (JITParticle or ScipyParticle)
                             lon      = release_lon,    # a vector of release longitudes 
                             lat      = release_lat,    # a vector of release latitudes
                             depth    = release_depth,  # a vector of release depths 
                             repeatdt = delta(hours=1)) # repeat release of particles
```

```python
kernels     = pset.Kernel(AdvectionRK4_3D) 
output_file = pset.ParticleFile(name="/ocean/brogalla/GEOTRACES/data/NEMO3D_fire_RK43D.nc", outputdt=delta(hours=1))
```

To add diffusion, you need to have a field of Kh_zonal

```python
# kernels2     = pset.Kernel(AdvectionDiffusionM1) #AdvectionDiffusionEM or AdvectionDiffusionM1
# output_file2 = pset.ParticleFile(name="/ocean/brogalla/GEOTRACES/data/NEMO3D_fire_ADEM.nc")
```

```python
pset.execute(kernels, runtime=delta(hours=40), dt=delta(hours=1), \
             recovery={ErrorCode.ErrorOutOfBounds: DeleteParticle}, \
             output_file=output_file)
output_file.export()
# dt      --- the timestep of the kernel
# runtime --- the total length of the run
```

## Look at particle tracks

```python
def load_tracks(filename):
    file  = xr.open_dataset(filename)
    traj  = file['trajectory'].values # dimensions: number of particles, tracks
    time  = file['time'].values       # units: seconds
    lat   = file['lat'].values        # degrees North
    lon   = file['lon'].values        # degrees East
    depth = file['z'].values        # degrees East

    return traj, time, lon, lat, depth
```

```python
traj_RK43D, time_RK43D, lon_RK43D, lat_RK43D, depth_RK43D = load_tracks('/ocean/brogalla/GEOTRACES/data/NEMO3D_fire_RK43D.nc')
```

```python
colors_time  = plt.cm.plasma(np.linspace(0,1,traj.shape[0]))
```

Color by time

```python
fig, ax1 = plt.subplots(figsize=(4.5, 3.74), dpi=300)

proj1 = Basemap(projection='stere',lat_0 = 48.5, lon_0 = -123.5, resolution='h', width=1e5, height=7e4, ax=ax1)
proj1.drawcoastlines(color='#373736', linewidth=1.0)
proj1.fillcontinents(color=land_color, lake_color=land_color, zorder=2);

x,y = proj1(lon_RK43D, lat_RK43D)

for particle in range(0,traj_RK43D.shape[0]):
    x_plot = x[particle,:][x[particle,:] < 1e10]
    y_plot = y[particle,:][y[particle,:] < 1e10]
    
    ax1.plot(x_plot, y_plot,'-', linewidth=0.7, color=colors_time[particle])
    ax1.plot(x_plot[-1], y_plot[-1],'.w', markersize=2, markeredgecolor='k')
    c = ax1.scatter([0,0],[0,0], s=0.3, c=[0,40], cmap='plasma_r', vmin=0, vmax=40) # just for colorbar
        
ax1.plot(x[0,0],y[0,0],'.r',markersize=10, markeredgecolor='w', markeredgewidth=1)
fig.colorbar(c, shrink=0.5, label='Time [hours]')
```

Color by depth

```python
fig, ax1 = plt.subplots(figsize=(4.5, 3.74), dpi=300)

proj1 = Basemap(projection='stere',lat_0 = 48.3, lon_0 = -123.5, resolution='h', width=1e5, height=9e4, ax=ax1)
proj1.drawcoastlines(color='#373736', linewidth=1.0)
proj1.fillcontinents(color=land_color, lake_color=land_color, zorder=2);

x,y = proj1(lon_RK43D, lat_RK43D)

for particle in range(0,traj_RK43D.shape[0]):
    x_plot = x[particle,:][x[particle,:] < 1e10]
    y_plot = y[particle,:][y[particle,:] < 1e10]
    z_plot = depth_RK43D[particle,:][~np.isnan(depth_RK43D[particle,:])]
    
    ax1.plot(x_plot, y_plot,'-k', linewidth=0.1)
    c = ax1.scatter(x_plot, y_plot, s=0.3, c=z_plot, cmap='plasma', vmin=0, vmax=50) #plasma
    
ax1.plot(x[0,0],y[0,0],'.r',markersize=10, markeredgecolor='w', markeredgewidth=1)
fig.colorbar(c, shrink=0.5, label='Depth [m]')
```
