#### Load data

```python
df = xr.open_dataset('/ocean/brogalla/GEOTRACES/data/IDP2021_GEOTRACES_IDP2021_Seawater_Discrete_Sample_Data_v1_Njx5WkTn.nc')
```

```python
dissolPb     = df['var9'].values  # dissolved Pb (pmol/kg)
# dissolPb_pmp = df['var11'].values # Concentration of dissolved Pb (pmol/kg)
# dissolPb_pmp = df['var12'].values # Concentration of dissolved Pb (pmol/kg) sub ice pump
particPb     = df['var15'].values # total particulate lead determined by filtration from a water sampling bottle (pmol/kg)
labilePb     = df['var17'].values # labile particulate lead determined by filtration from a water sampling bottle (pmol/kg)
particPb_LPT = df['var19'].values # total particulate lead by in situ filtration (pump) collected on a prefilter (large particles); pmol/kg
particPb_SPT = df['var20'].values # same but small particles
POC_LPT      = df['var21'].values # Concentration of POC by in situ filtration (pump) collected on a prefilter (large particles) (umol C/kg)
POC_SPT      = df['var22'].values # same but small particles
pmass_LPT    = df['var23'].values # Concentration of particulate mass (dry weight) (ug/kg) determined by in situ filtration (pump) collected on a prefilter (large particles)
pmass_SPT    = df['var24'].values # same as above but for small particles
```
