## Look at number of particles over time

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import glob 
%matplotlib inline
```

```python
# domain dimensions:
jmin, jmax = 159, 799
imin, imax = 1139, 2179
```

```python
folder = '/ocean/brogalla/GEOTRACES/data/Pb-tuning-202302/fsb-b010-p015-m5/'

files = np.sort(glob.glob(f'{folder}Pb_scavenging_y20*'))

part = []
for file in files:
    print(file)
    dset = xr.open_dataset(file)
    part.append(np.sum(dset['particles'].values[:,imin:imax,jmin:jmax], axis=(0,1,2)))
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5), dpi=300)
ax.plot(part)
# ax.plot([0,250], [0, 1], '--')
ax.set_xlabel('time')
ax.set_ylabel('number of particles')
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5), dpi=300)
ax.plot(np.cumsum(part)/np.amax(np.cumsum(part)))
ax.plot([0,250], [0, 1], '--')
ax.set_xlabel('time')
ax.set_ylabel('number of particles')
```
