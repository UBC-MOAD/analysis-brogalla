# Animation of monthly files:

```python
rc('animation', html='html5')
```

```python
def plot_mn(frames):
    
    x, y = proj1(lon_model, lat_model)
    n    = int(np.floor(frames/12)) # year
    i    = int(frames-(n*12)) # month within year
    
    cbar = proj1.pcolor(x[::5,::5], y[::5,::5], mn_dissolved[i,n,depth,::5,::5],\
                        vmin=min_v, vmax=max_v, cmap=colour)
    ax1.set_title('Year: '+str(2002+n)+', month: '+ str(i+1))
    
    cbar_ax = fig.add_axes([0.91, 0.15, 0.04, 0.7])
    fig.colorbar(cbar, cax=cbar_ax)

    return 
```

```python
fig, ax1, proj1 = pickle.load(open('/ocean/brogalla/GEOTRACES/analysis-brogalla/maps/mn-reference.pickle','rb'))

# Run animation:
ani = animation.FuncAnimation(fig, plot_mn, frames=12*14, interval= 12*13*5)
ani.save('/ocean/brogalla/GEOTRACES/figures/animation-reference-run2.mp4')
```
