```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

%matplotlib notebook
```

Kadko et al. (2019) estimate Pb residence times in the Arctic Ocean and TPD. Let's see what we can learn from this.

```python
station = [14,19,26,30,31,32,38,43,46,48,51,52,53,54,57,60] # station number
T_Pb    = np.array([0.69, 0.76, 0.94, 0.24, 0.76, 0.27, 0.33, 0.71, 1.07, 0.65, 0.31, 0.36, \
                    0.41, 0.28, 0.38, 0.18]) # yrs
stn_lon = np.array([-173.034, -174.962, 174.915, -179.809, -176.761, -89.25, -149.734, -150.063, \
                    -149.872, -149.855, -147.831, -148.009, -148.803, -149.515, -156.808, -158.729])
stn_lat = np.array([76.511, 79.997, 83.755, 87.520, 88.407, 89.987, 87.815, 85.135, 82.491, 80.369, 78.180, \
                    77.503, 76.998, 76.505, 73.506, 73.008])
```

Map out the distribution of residence times:

```python
fig, ax1, proj1 = pickle.load(open('/ocean/brogalla/GEOTRACES/pickles/AO_map.pickle','rb'))

# Plot residence time estimates -------------------------------------------------------------
x_stn, y_stn  = proj1(stn_lon, stn_lat) 
res_time = proj1.scatter(x_stn, y_stn, c=T_Pb, s=120, edgecolor='w')


# Color bar: --------------------------------------------------------------------------------
cbaxes = fig.add_axes([0.2, 0.22, 0.3, 0.03]);
CBar   = plt.colorbar(res_time, ax=ax1, cax=cbaxes, orientation='horizontal') 
CBar.set_label('Residence time of Pb in upper ocean [years]', fontsize=10)
CBar.ax.tick_params(axis='y', length=0, labelsize=10)
```
